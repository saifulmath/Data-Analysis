                                        Grouping Data by Continent

```python
print(df['Continent'].unique())
```

```python
type(df['Continent'][0])
```

```python
df.groupby('Continent').count()
```

```python
#To calculate the mean for numeric columns, we can select the numeric columns before grouping
numeric_columns = df.select_dtypes(include='number').columns
result = df.groupby('Continent')[numeric_columns].mean()
result
```

```python
result1 = df.groupby('Continent')[numeric_columns].mean().sort_values(by="2022 Population",ascending=False)
result1
```

```python
result2 = df.groupby('Continent')[numeric_columns].sum().sort_values(by="2022 Population",ascending=False)
result2
```

```python
df[df['Continent'] == 'Oceania']
```

                                          Visualizing Grouped Data 

```python
#plt.figure(figsize = (22,6))
result1.plot(figsize = (20,8))
```

```python
result3 = result1.transpose()
result3
```

```python
result3.plot(figsize = (20,6))
#plt.figure(figsize = (22,6))
```

```python
df.columns
```

```python
result4 = df.groupby('Continent')[df.columns[5:13]].mean().sort_values(by="2022 Population",ascending=False)
result4
```

```python
columns_reordered = ['1970 Population','1980 Population','1990 Population','2000 Population','2010 Population','2015 Population','2020 Population','2022 Population']
```

```python
result4_reordered = result4[columns_reordered]
```

```python
result4_reordered
```

```python
result5 = result4_reordered.transpose()
result5
```

```python
result5.plot(figsize = (18,8))
plt.xlabel('Year',fontsize = 14, fontweight = "bold")
plt.ylabel('Population (mean)',fontsize = 14, fontweight = "bold")
plt.title('World Population Growth',fontsize = 14, fontweight = "bold")
```
